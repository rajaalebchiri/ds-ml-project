```python
import csv
import numpy as np
```

### Slicing NumPy Arrays

```python
with open('dataset.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
print(data)
```

```python
numpy_arr = np.array(data)
```

```python
numpy_arr[1,2]
```

```python
# get one column
numpy_arr[:,1]
```

```python
numpy_arr[1, :]
```

```python
slice = numpy_arr[2]
```

```python
slice
```

```python
numpy_arr[2]
```

```python
slice[3] = 1000
```

```python
slice
```

```python
numpy_arr[2]
```

```python
# partial slicing
numpy_arr[1, 1:6]
```

```python
numpy_arr[100:103,5:]
```

```python
numpy_arr[105:-1, 2:3]
```

```python
numpy_arr[105:-1:5, 2:3]
```

```python
numpy_arr[2:25:3,:]
```

```python
numpy_arr[1:8:2,:]
```

### Indexing, Filtering, and Broadcasting NumPy Arrays

```python
numpy_dataset = np.genfromtxt('dataset.csv', delimiter=',')
```

```python
numpy_dataset
```

```python
numpy_dataset[3, 4]
```

```python
indexer_tuple = (3,4)
```

```python
numpy_dataset[indexer_tuple]
```

advanced indexing using mutable data structures

```python
indexer_arr = [3, 4]
```

```python
numpy_dataset[indexer_arr]
```

masking

```python
small_arr = numpy_dataset[:3, :3]
```

```python
small_arr
```

```python
# mask to find values that are greater than 2
mask = np.where(small_arr > 2)
```

```python
mask
```

```python
small_arr[mask]
```

Broadcasting

```python
np.array(range(6))
```

```python
np.array(6) * np.array(range(6))
```

```python
2 + np.zeros(5)
```

```python
a = np.arange(5).reshape(1,5)
a
```

```python
b = np.arange(25).reshape(5,5)
b
```

```python
a * b
```
